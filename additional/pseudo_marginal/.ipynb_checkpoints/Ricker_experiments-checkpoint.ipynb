#### Testing segmented particle MCMC for hidden Markov models:

```python
# Calling libraries:
from __future__ import division
%matplotlib inline
import numpy as np, time, matplotlib.pyplot as plt, math, pandas, numpy.random as npr, scipy
from pylab import plot, show, legend
from time import time
from scipy.stats import *
from Ricker_functions import *
```

```python
x_0 = 5; T = 200
r, sigma, phi = np.exp(2.8), 0.5, 3
theta = np.asarray([r, sigma, phi])

X, Y = simulate_data(x_0, T, theta)
```

### Run segmented MCMC:

```python
N_MCMC = 5_000
N_PF = 200
scale = [2e-1, 2e-1, 2e-1]
chunk_size = 50
power = 1

start = time()
θ_chains_apprx = chunked_pMCMC(x_0, Y, theta, N_MCMC, scale, N_PF, chunk_size, power)
print(round((time()-start)/60, 2), 'minutes')
```

#### Density estimates for chunk posteriors:

```python
params = [r"$log \, r$", r"$log \, \sigma$", r"$log \, \phi$"]
plt.figure(figsize=(20,8))

thin = 10
for j in range(3) :
    for i in range(np.shape(θ_chains_true)[0]) :
        plt.subplot(3,4,4*j+i+1) 
        x_vals = np.linspace(start=np.min(np.log(θ_chains_true[i,::thin,j])), 
                                 stop=np.max(np.log(θ_chains_true[i,::thin,j])), num=100)
        plt.plot(x_vals, gaussian_kde(np.log(θ_chains_true[i,::thin,j])) (x_vals), "red", 
                 label="true subset posterior") 
        x_vals = np.linspace(start=np.min(np.log(θ_chains_apprx[i,::thin,j])), 
                                 stop=np.max(np.log(θ_chains_apprx[i,::thin,j])), num=100)
        plt.plot(x_vals, gaussian_kde(np.log(θ_chains_apprx[i,::thin,j])) (x_vals), "blue",
                label="appproximate subset posterior (through buffering)") 
        plt.grid(True)
        if i == 0 : plt.ylabel(params[j], fontsize=15)
        if j == 0 : plt.title("Chunk "+str(i+1))
    if j == 0 : plt.legend(ncol=2, bbox_to_anchor=(0,1.6), fontsize=14, frameon=False)
plt.suptitle("Density estimates for chunk posteriors", x=0.5, y=0.95, fontsize=17);
```

#### Traceplots for chunk posteriors:

```python
for j in range(3) :
    plt.figure(figsize=(20,2))
    for i in range(np.shape(θ_chains_true)[0]) :
        plt.subplot(1,4,1+i) 
        plt.plot(np.log(θ_chains_true[i,:,j]))
        plt.grid(True)
        plt.suptitle("True", fontsize=15)
        plt.title("Chunk "+str(i+1))
        if i == 0 : plt.ylabel(params[j], fontsize=12)
    plt.show()
    
for j in range(3) :
    plt.figure(figsize=(20,2))
    for i in range(np.shape(θ_chains_apprx)[0]) :
        plt.subplot(1,4,1+i) 
        plt.plot(np.log(θ_chains_apprx[i,:,j]))
        plt.grid(True)
        plt.suptitle("Approximate", fontsize=15)
        plt.title("Chunk "+str(i+1))
        if i == 0 : plt.ylabel(params[j], fontsize=12)
    plt.show()
```

### Run full MCMC:

```python
N_PF = 500
scale = [0.075,0.075,0.075]

start = time()
θ_chain_full = pMCMC(x_0, Y, θ, N_PF, N_MCMC, scale)
print( round((time() - start)/60, 2), 'minutes' )
```

```python
params = [r"$log \, r$", r"$log \, \sigma$", r"$log \, \phi$"]
plt.figure(figsize=(20,3))

thin = 10
for i in range(3) :
    plt.subplot(1,3,i+1) 
    x_vals = np.linspace(start=np.min(np.log(θ_chain_full[::thin,i])), 
                             stop=np.max(np.log(θ_chain_full[::thin,i])), num=100)
    plt.plot(x_vals, gaussian_kde(np.log(θ_chain_full[::thin,i])) (x_vals)) 
    plt.grid(True)
    plt.title(params[i], fontsize=16)
plt.suptitle("Density estimates for posteriors", x=0.5, y=1.1, fontsize=17);
```

```python
plt.figure(figsize=(20,3))
thin = 10
for i in range(3) :
    plt.subplot(1,3,i+1) 
    plt.plot(np.log(θ_chain_full[:,i]))
    plt.grid(True)
    plt.title(params[i], fontsize=16)
plt.suptitle("Traceplots", x=0.5, y=1.1, fontsize=17);
```

```python
full_mean = np.mean(np.log(θ_chain_full),0)
apprx_mean = np.mean([np.mean(np.log(θ_chains_apprx[i]),0) for i in range(3)],0)
true_mean = np.mean([np.mean(np.log(θ_chains_true[i]),0) for i in range(3)],0)
```

```python
scipy.linalg.norm(full_mean-apprx_mean), scipy.linalg.norm(full_mean-true_mean)
```
