### Divide-and-conquer MCMC:

```python
n_mcmc = 10_000
n_particles = T
m = 10
power = m
scale = sc/np.sqrt((T/(tend-tstart)))

tstarts = np.arange(m).astype(int)
tends = 1 + tstarts
tstarts *= int(T/m)
tends *= int(T/m)
def f(y) :
    return pMCMC(x_0, y, theta_0, n_particles, n_mcmc, scale, power, adapt=True, start_adapt=0.2)[0]
Ychunks = [Y[tstart:(int(T/m)+tstart)] for tstart in tstarts]

pool = mp.Pool(min(m,10))
results_m10 = pool.map(f, [y for y in Ychunks])
pool.close()
```

```python
file = open("/xtmp/DC-BATS_data/HMMs/ricker/mcmc_m10.pkl", "wb")
pickle.dump(results_m10, file)
file.close()
```

### Compare divide-and-conquer with full MCMC:

```python
mean_dc_m5, mean_dc_m10, mean_dc_m20 = np.zeros(3), np.zeros(3), np.zeros(3)
var_dc_m5, var_dc_m10, var_dc_m20 = np.zeros(3), np.zeros(3), np.zeros(3)
median_dc_m5, median_dc_m10, median_dc_m20 = np.zeros(3), np.zeros(3), np.zeros(3)
ci_dc_m5, ci_dc_m10, ci_dc_m20 = np.zeros((2,3)), np.zeros((2,3)), np.zeros((2,3))
for chunk in range(5) :
    mean_dc_m5 += np.mean(results_m5[chunk],0)/5
    var_dc_m5 += np.var(results_m5[chunk],0)/5
    median_dc_m5 += np.median(results_m5[chunk],0)/5
    ci_dc_m5 += np.percentile(results_m5[chunk],axis=0,q=[2.5,97.5])/5
for chunk in range(10) :
    mean_dc_m10 += np.mean(results_m10[chunk],0)/10
    var_dc_m10 += np.var(results_m10[chunk],0)/10
    median_dc_m10 += np.median(results_m10[chunk],0)/10
    ci_dc_m10 += np.percentile(results_m10[chunk],axis=0,q=[2.5,97.5])/10
for chunk in range(20) :
    mean_dc_m20 += np.mean(results_m20[chunk],0)/20
    var_dc_m20 += np.var(results_m20[chunk],0)/20
    median_dc_m20 += np.median(results_m20[chunk],0)/20
    ci_dc_m20 += np.percentile(results_m20[chunk],axis=0,q=[2.5,97.5])/20
ci_full = np.percentile(theta_chain_full,axis=0,q=[2.5,97.5])
var_full = np.var(theta_chain_full, axis=0)
```
