### Model is:

\begin{eqnarray}
X_1 
& \sim & 
\mathcal{N}(0,1) 
\\
X_t 
& = & 
\phi X_{t-1} + \mathcal{N}(0, \sigma^2)
\\
Y_t 
& \sim &
\mathcal{N}(0, \tau^2 e^{X_t}).
\end{eqnarray}

```python
# Calling libraries:
from __future__ import division
%matplotlib inline
import numpy as np, time, matplotlib.pyplot as plt, math, pandas, numpy.random as npr, multiprocessing as mp, scipy, gc
from pylab import plot, show, legend
from time import time
from scipy.stats import *
from SVM import *

gc.enable()
```

```python
phi = 0.3
sigma2 = 1
tau2 = 1
theta = (phi, sigma2, tau2)

T = 50
X, Y = simulate_data(theta, T)
```

### Plot likelihoods:

```python
I = 100
n_particles = 500
phi_grid = np.linspace(-0.9,0.95,I)
sigma2_grid = np.exp(np.linspace(-2.5,2.5,I))
ll_grid = np.zeros((I,I))

for i1 in trange(I) :
    for i2 in range(I) :
        theta_test = np.asarray([phi_grid[i1], sigma2_grid[i2], tau2])
        ll_grid[i1,i2] = bootstrap_PF_mult(n_particles, theta_test, Y, m=1)
```

```python
plt.imshow(ll_grid[:I,:I]);
```

### Run MCMC:

```python
n_particles = 2*T 
n_mcmc = 10_000
sc = [0.35, 0.93, 0.67]
theta_t = [theta[0], np.log(theta[1]), np.log(theta[2])]

titles = [r"$\phi$", r"$\log \, \sigma^2$", r"$\log \, \tau^2$"]
plt.rcParams['figure.figsize'] = (16, 3)
```

```python
power = 1
scale = sc/np.sqrt(power)
theta_chain_p1, scales_p1 = pMCMC(Y, theta, n_particles, n_mcmc, scale, power=power)

for i in range(3) :
    plt.subplot(1,3,i+1)
    plt.plot(np.arange(1+n_mcmc), theta_chain_p1[:,i])
    plt.axhline(y=theta_t[i], color="red")
    plt.grid(True)
    plt.title(titles[i], fontsize=14)
    plt.xlabel("MCMC iteration")
```

```python
power = 5
scale = sc/np.sqrt(power)
theta_chain_p5, scales_p5 = pMCMC(Y, theta, n_particles, n_mcmc, scale, power=power)

for i in range(3) :
    plt.subplot(1,3,i+1)
    plt.axhline(y=theta_t[i], color="red")
    plt.plot(np.arange(1+n_mcmc), theta_chain_p5[:,i])
    plt.grid(True)
    plt.title(titles[i], fontsize=14)
    plt.xlabel("MCMC iteration")
```

```python
power = 10
scale = sc/np.sqrt(power)
theta_chain_p10, scales_p10 = pMCMC(Y, theta, n_particles, n_mcmc, scale, power=power)

for i in range(3) :
    plt.subplot(1,3,i+1)
    plt.axhline(y=theta_t[i], color="red")
    plt.plot(np.arange(1+n_mcmc), theta_chain_p10[:,i])
    plt.grid(True)
    plt.title(titles[i], fontsize=14)
    plt.xlabel("MCMC iteration")
```

```python
burnin = 2000
print("Pow = 5 : ", np.var(theta_chain_p1[burnin:],0)/np.var(theta_chain_p5[burnin:],0))
print("Pow = 10: ", np.var(theta_chain_p1[burnin:],0)/np.var(theta_chain_p10[burnin:],0))
print("Pow = 5 : ", np.var(np.exp(theta_chain_p1)[burnin:],0)/np.var(np.exp(theta_chain_p5)[burnin:],0))
print("Pow = 10: ", np.var(np.exp(theta_chain_p1)[burnin:],0)/np.var(np.exp(theta_chain_p10)[burnin:],0))
```
