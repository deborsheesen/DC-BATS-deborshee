### Random walk Metropolis algorithm

#### Model: 

\begin{eqnarray}
\text{Latent:} \quad X_t & = & A X_{t-1}  + \nu_t, 
\\
\text{Observed:} \quad Y_t & = & C X_t + B Z_t + \omega_t, \quad Z_t ~ \text{are covariates}
\\
\nu_t & \sim & \text{N}(0, Q ),
\\
\omega_t & \sim & \text{N}(0, R).
\end{eqnarray}
