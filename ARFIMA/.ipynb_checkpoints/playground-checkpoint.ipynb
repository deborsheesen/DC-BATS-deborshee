```python
# Calling libraries:
from __future__ import division
%matplotlib inline
import numpy as np, time, matplotlib.pyplot as plt, math, pandas, numpy.random as npr, pystan as ps, pickle
from pylab import plot, show, legend
from time import time
from tqdm import trange
from scipy.stats import *
```

Model is
$$ 
\left ( 1 - \sum_{i=1}^p \phi_i B^i \right ) y_t 
=
\left ( 1 + \sum_{j=1}^q \theta_j B^j \right ) (1-B)^{-d} \varepsilon_t.
$$

Now let $Z_t = (1-B)^{-d} \varepsilon_t$

\begin{align}
y_t 
& = 
\beta^\top X_t + \sum_{i=1}^p \phi_i y_{t-i} + Z_t + \sum_{j=1}^q \theta_q Z_{t-j}
\\
Z_t 
& = 
\sum_{s=0}^{t} b_s \varepsilon_{t-s}, \quad \varepsilon_s \stackrel{\text{i.i.d.}}{\sim} \mathrm{N}(0,1)
\end{align}

```python
d = 1/3
T = 2*10**4
eps = npr.randn(T)
b = np.zeros(T)
for j in np.arange(0,150) :
    b[j] = math.gamma(j+d)/(math.gamma(j+1)*math.gamma(d))
for j in np.arange(150,T) :
    b[j] = j**(d-1)/math.gamma(d)
```

```python
Z = np.zeros(T)
for t in trange(T) :
    Z[t] = np.sum(b[:t]*(eps[:t][::-1]))
```

```python
p = 3
q = 3
r = max(p,q)
phi = npr.randn(p)/4
theta = npr.randn(q)/4

d = 2
beta = npr.randn(d)
X = npr.rand(d,T)

y = np.zeros(T)
for t in np.arange(0,r) :
    y[t] = beta.dot(X[:,t]) + npr.randn()
for t in np.arange(r,T) :
    y[t] = beta.dot(X[:,t]) + np.sum(phi*y[(t-p):t]) + Z[t] + np.sum(theta*Z[(t-q):t])
```

```python
plt.figure(figsize=(18,4))
plt.plot(y)
plt.xlim(-0.005*T,1.005*T)
plt.title("Observations", fontsize=14)
plt.xlabel("Time index", fontsize=13)
plt.grid(True);
```

```python
def acf(y, lag) :
    T = len(y)
    y1 = y[0:(T-lag)]
    y2 = y[lag:T]
    return (np.mean(y1*y2) - np.mean(y1)*np.mean(y2))/(np.std(y1)*np.std(y2))
```

```python
maxlag = 500
acfs_y = np.zeros(maxlag)
for lag in range(maxlag) :
    acfs_y[lag] = acf(y,lag)
plt.plot(acfs_y)
plt.grid(True)
plt.ylim(0,1)
```
